```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3]
y = [2, 4, 6]

# Create a line plot
plt.plot(x, y)
```

```python
# only show the points
plt.scatter(x, y)
```

```python
# coloring
plt.plot(x, y, linestyle='--', color='red')
```

```python
plt.scatter(x, y, marker='x', color='green')
```
